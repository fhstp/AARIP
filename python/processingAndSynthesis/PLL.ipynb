```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
```

```python
from IPython.display import Audio
```

```python
x,sr = librosa.load('/root/Documents/MATLAB/data/drumLoop.wav', sr = None)
N = len(x)
n = np.arange(N)
t = n/sr
```

```python
plt.specgram(x,NFFT=2**12, noverlap=2**12-100, Fs=sr, cmap= 'gnuplot2', vmin=-100)
plt.show()
```

```python
fc = 1000
carrier = np.cos(2*np.pi*t*fc)
```

```python
plt.specgram(carrier,Fs=sr, cmap= 'gnuplot2', vmin=-100)
plt.show()
```

```python
modulated = carrier*x
```

```python
plt.specgram(modulated,Fs=sr, cmap= 'gnuplot2', vmin=-100)
plt.show()
```

```python
Audio(data=modulated, rate = sr)
```
